# Wielowymiarowy rozkład normalny

```python
import matplotlib.pylab as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.mlab as mlab
from matplotlib import cm
from scipy.stats import multivariate_normal
import seaborn as sns
```

# Zad. 
Narysujmy gęstość dwuwymiarowego rozkładu Normalnego o parametach 

* n=[0,0], cov = [[1,0],[0,1]]
* n=[0,0], cov = [[2,0],[0,1]]
* n=[0,0], cov = [[1,0],[0,2]]

```python
def drawDensity(n, cov):
    
    x=np.arange(-5,5,0.1)
    y=x
    
    X,Y = np.meshgrid(x,y)
    pos = np.dstack((X,Y))
    print(np.shape(pos))
    rv = multivariate_normal(mean=n, cov=cov)
    Z = rv.pdf(pos)
    
    print(np.shape(Z))
    
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    ax.plot_surface(X,Y,Z, cmap="jet")
    plt.show()
    
drawDensity([0,0], [[1,0],[0,1]])
drawDensity([0,0], [[2,0],[0,1]])
drawDensity([0,0], [[1,0],[0,2]])
```

# Zad 
Narysujmy kontury gęstości dwuwymiarowego rozkładu Normalnego o parametach 

* n=[0,0], cov = [[1,0],[0,1]]
* n=[0,0], cov = [[2,0],[0,1]]
* n=[0,0], cov = [[1,0],[0,2]]

```python
def drawContours(n,cov):
    x = np.linspace(-10,10,1000)
    y = np.linspace(-10,10,1000)
    
    X,Y = np.meshgrid(x,y)
    pos = np.dstack((X,Y))
    rv = multivariate_normal(mean=n,cov=cov)
    
    Z = rv.pdf(pos)
    
    fig = plt.figure()

    ax = fig.add_subplot()
    ax.set_xlim(-2.5,2.5)
    ax.set_ylim(-2.5,2.5)
    ax.contour(X,Y,Z,50)
    
    plt.show()
    
drawContours([0,0], [[1,0],[0,1]])
drawContours([0,0], [[2,0],[0,1]])
drawContours([0,0], [[1,0],[0,2]])
```

# Zad 
Wylosujmy próbkę z dwuwymiarowego rozkładu Normalnego o parametach 

* n=[0,0], cov = [[1,0],[0,1]]
* n=[0,0], cov = [[2,0],[0,1]]
* n=[0,0], cov = [[1,0],[0,2]]

i narysuj odpowiedni wykres

```python
def plotScatter(n, cov):
    rv = multivariate_normal.rvs(mean=n,cov=cov, size=1000)
    plt.scatter(rv[:,0], rv[:,1])
    plt.show()
    
plotScatter([0,0], [[1,0],[0,1]])
plotScatter([0,0], [[2,0],[0,1]])
plotScatter([0,0], [[1,0],[0,2]])
```
